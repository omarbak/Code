```python
from sklearn.utils import resample
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from collections import Counter
sns.set()
```

```python
def plot_data(X, y,alpha=0.5):
    plt.scatter(X[y == 0, 0], X[y == 0, 1], label="Class #0", alpha=alpha, linewidth=0.15)
    plt.scatter(X[y == 1, 0], X[y == 1, 1], label="Class #1", alpha=alpha, linewidth=0.15, c='orange')
    plt.legend()
    return plt.show()
```

## Creating a Dataset.
- Data is noise free.
- 5% positive class only
- minimal sub-classes
- 2 features

```python
X, y = make_classification(n_samples=500,n_features=2,n_classes=2,n_clusters_per_class=1,
                           weights=[0.95],n_redundant=0,random_state=0)
```

```python
counter = Counter(y)
```

```python
# visualizing the original data
plt.figure(figsize=(14,7)) 
plt.title("Original Data distribution")
plot_data(X,y,alpha=0.8)
```

```python
X0 = pd.Series(X[:,0])
X1 = pd.Series(X[:,1])
y = pd.Series(y)
```

```python
dataset = pd.concat([X0,X1,y],axis=1)
dataset.head()
```

```python
dataset.columns = ["X0","X1","Class"]
```

```python
dataset.Class.value_counts()
```

```python
majority = dataset[dataset.Class==0]
minority = dataset[dataset.Class==1]
```

```python
minority.head()
```

#### Upsampling the minority class.
- Make the minority class equal to the majority class by resampling with replacement aka duplicating instances.
- Each example will be occuring multiple times in the dataset. 
- The up-sampling is random, therefore, not all duplicated examples with have exact amount.

```python
# Upsample minority class
minority_upsampling = resample(minority, 
                                 replace=True,     # we have to use replacement since we are upsampling
                                 n_samples=len(majority),    # as much as majority
                                 random_state=0)
```

```python
upsampled_minority_df = pd.concat([majority,minority_upsampling])
```

```python
upsampled_minority_df.Class.value_counts()
```

```python
upsampled_minority_df.tail()
```

```python
X_upsampled = np.array(pd.concat([upsampled_minority_df['X0'],upsampled_minority_df['X1']],axis=1))
```

```python
counter_upsampled = Counter(upsampled_minority_df.Class)
```

##### Plotting the data.
- Since we did an upsampling, the same point will have multiple duplicates.
- We lower the alpha in the scatter plot to see the effect or density of colors for the new dataset.
- Duplicated points will have a stronger color/density than non duplicates.

```python
plt.figure(figsize=(14,7)) 
for label, _ in counter_upsampled.items():
    row_ix = np.where(upsampled_minority_df.Class == label)[0]
    plt.scatter(X_upsampled[row_ix,0],X_upsampled[row_ix,1], label=str(label),alpha=0.3)
    plt.legend()
# show the figure
plt.title("Upsampled data")
plt.show()
```

```python
plt.figure(figsize=(10,5)) 
for label, _ in counter_upsampled.items():
    row_ix = np.where(upsampled_minority_df.Class == label)[0]
    plt.hist2d(X_upsampled[row_ix,0],X_upsampled[row_ix,1],(50,50),cmap=plt.cm.jet)
    plt.colorbar()
# show the figure
plt.title("Upsampled data Density distribution")
plt.show()
```

```python
dups = upsampled_minority_df.pivot_table(index = ['X0','X1'], aggfunc ='size') 
above_1 = dups[dups>1]
```

```python
counts_df = pd.DataFrame(above_1,columns=["Count"])
```

```python
couples_list = ['C{}'.format(i) for i in range(len(counts_df))]
```

```python
counts_df['Couple'] = couples_list
```

```python
plt.figure(figsize=(20,10)) 
sns.barplot(x=counts_df.Couple,y=counts_df.Count)
plt.title("Duplicate counts for each couple(Upsampling)")
plt.show()
```

#### Downsampling the majority class.
- We will pick random points from the data to draw them.
- Data could be highly unrepresentative of the original distribution.

```python
majority_downsampling = resample(majority, 
                                 replace=False,     # unique samples
                                 n_samples=len(minority),    # as much as minority
                                 random_state=0)
```

```python
downsampled_majority_df = pd.concat([majority_downsampling,minority])
downsampled_majority_df.Class.value_counts()
```

```python
X_downsampled = np.array(pd.concat([downsampled_majority_df['X0'],downsampled_majority_df['X1']],axis=1))
```

```python
counter_downsampled = Counter(downsampled_majority_df.Class)
```

```python
plt.figure(figsize=(14,7)) 
for label, _ in counter_downsampled.items():
    row_ix = np.where(downsampled_majority_df.Class == label)[0]
    plt.scatter(X_downsampled[row_ix,0],X_downsampled[row_ix,1], label=str(label),alpha=0.8)
    plt.legend()
# show the figure
plt.title("Downsampled Majority group")
plt.show()
```

### Avanced sampling methods.
- Synthetic Sampling (SMOTE)
- Adaptive Synthetic sampling.
- We are going to try multiple algorithms and highlight the difference between them.

```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import BorderlineSMOTE,SVMSMOTE,ADASYN,SMOTENC,RandomOverSampler
```

```python
# regular smote method; all the points will be resampled with neighbors.
def perform_SMOTE(neighbors=5):
    method = SMOTE(random_state=0,k_neighbors=neighbors)

    # Create the resampled feature set
    X_resampled, y_resampled = method.fit_sample(X, y)
    print("Data Distribution is: ",Counter(y_resampled))
    # Plot the resampled data
    return plot_data(X_resampled, y_resampled)
```

#### SMOTE.
- creating artificial data using the surrouding similar points.
- Changing the K in K-Nearest points for the algorithm can produce different results.
    - For example choosing K = 3 we see that the synthetic points are tighter (closer to each others).
    - For K = 10 the coverage of synthetic data points gets wider.
    - Desntiy for higher K is lower, however, not recommended to get too high with neighbors since it is possible to create an un representative dataset.
- Difference between SMOTE and random over-sampling is that new fake data similar to the original one is generated.

\begin{align}
x_{new} & = \delta (\dot{x}_i-x_i) + x_i \\
\end{align}



```python
for k in [3,5,8,10]:
    plt.title("SMOTE {} Neighbors".format(k))
    perform_SMOTE(k)
```

#### BorderLine-SMOTE.
- Improved version of SMOTE.
- The points that are near the "DANGER" zone are the ones resampled.
- DANGER zone is defined by the data points from the minority class that are near the majority class (Hard points to classify).
- If a datapoint from the minority class is surrounded by M - majority class neighbors then this point is considered as noise and won't be synthesized. 
- In other words, easy to classify points are not resampled, noisy considered points are not resampled.
    - higher neighbors can lead to data disturbance and unrepresentitiveness
    - higher neighbors can lead to disturb the reason behind the usage of Borderline
- BorderLine2 from an observational point seems to create data that is relatively closer to the cluster of the second data class. Which means, that it tries to focus on the bordering line between the class to learn the hardest points possible.

```python
def perform_borderlineSMOTE(neighbors=5,kind='borderline-1'):
    # initiate model
    method = BorderlineSMOTE(random_state=0,k_neighbors=neighbors,kind=kind)
    # Create the resampled feature set
    X_resampled, y_resampled = method.fit_sample(X, y)
    print("Data Distribution is: ",Counter(y_resampled))
    # Plot the resampled data
    return plot_data(X_resampled, y_resampled)


for k in [3,5,8,10]:
    plt.title("BorderlineSMOTE {} Neighbors".format(k))
    perform_borderlineSMOTE(k)
for k in [3,5,8,10]:
    plt.title("BorderlineSMOTE(2) {} Neighbors".format(k))
    perform_borderlineSMOTE(k,kind='borderline-2')
```

#### ADASYN.
- the newly generated data points will be according to the distribution of points.
- Points will have a more equal distribution compared to previous algorithms.

```python
def perform_ADASYN(neighbors=5):
    # initiate model
    method = ADASYN(random_state=0,n_neighbors=neighbors)
    # Create the resampled feature set
    X_resampled, y_resampled = method.fit_sample(X, y)
    print("Data Distribution is: ",Counter(y_resampled))
    # Plot the resampled data
    return plot_data(X_resampled, y_resampled)

for k in [3,5,8,10]:
    plt.title("Adaptive Synthetic sampling {} Neighbors".format(k))
    perform_ADASYN(k)
```

### Creating a Model.
- We are going to create a logistic regression model on the basic data and see if would produce good results.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
np.random.seed(0)
```

```python
from imblearn.pipeline import Pipeline
```

```python
X_new, y_new = make_classification(n_samples=2000,n_features=2,n_classes=2,n_clusters_per_class=1,
                           weights=[0.97],n_redundant=0,random_state=11)
```

```python
plot_data(X_new,y_new)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_new,y_new,test_size=0.25,random_state=0)
```

```python
Counter(y_test)
```

```python
## Checking if the classes are balanced as the original distribution.
print("Class 1 fraction train", Counter(y_train)[1]/len(y_train))
print("Class 1 fraction test", Counter(y_test)[1]/len(y_test))
```

##### Baseline Model.

```python
# Initiating and training the model.
model = LogisticRegression(random_state=0)
# train and fit
model.fit(X_train,y_train)

# model predictions
predicted = model.predict(X_test)
# Print the classifcation report and confusion matrix
print('Classification report:\n', classification_report(y_test, predicted))

conf_mat = confusion_matrix(y_true=y_test, y_pred=predicted)
sns.heatmap(pd.DataFrame(conf_mat,index=["False","True"],columns=["False","True"]),annot=True,fmt='d')
plt.title("Confusion Matrix")
plt.show()
```

##### Logistic Regression + Equalized Upsampling.
- In the following we try multiple upsampling algorithms with a logistic regression model and we check the different metrics.
- Equal upsampling is not 'the' optimal upsampling, however, we are going to use it.
- Later work, is the use of Upsampling and downsampling methods. on a similar model.

```python
def draw_confusion(confusion_matrix,model_name):
    sns.heatmap(pd.DataFrame(confusion_matrix,index=["Act.False","Act.True"],columns=["Pred.False","Pred.True"]),annot=True,fmt='d')
    plt.title("Confusion Matrix of {}".format(model_name))
    plt.show()
    
def imb_pipeline(model_name,model_instantiation,classifier_name="Logistic Regression",classifier=LogisticRegression(random_state=0),X=X_train,y=y_train,x_test=X_test,y_true = y_test):
    # pipeline containing the resampling method and the LogReg
    pipeline = Pipeline([(model_name,model_instantiation),(classifier_name,classifier)])
    # training and fitting the model
    pipeline.fit(X,y)
    # predicting on the test set
    predicted_pipeline = pipeline.predict(x_test)
    # report
    print('Classification report for {}:\n'.format(model_name), classification_report(y_true, predicted_pipeline))
    #confus matrix
    conf_mat = confusion_matrix(y_true, predicted_pipeline)
    draw_confusion(conf_mat,model_name)
```

```python
# # GENERAL FORM
dictionary_model = {"Random":RandomOverSampler(random_state=0),
                    "SMOTE":SMOTE(random_state=0),
                    "Border":BorderlineSMOTE(random_state=0),
                    "ADAPT":ADASYN(random_state=0)}

for model, model_instantiation in dictionary_model.items():
    # resampling the data
    X_oversampled,y_oversampled = model_instantiation.fit_sample(X_train,y_train)
    # plotting the oversampled data
    plt.title("Oversampling using {}".format(model))
    plot_data(X_oversampled,y_oversampled)
    # Pipeline run.
    imb_pipeline(model,model_instantiation)
```

##### Logistic Regression + Upsampling-Data Cleaning (SMOTE+ENN, Tomek Links+SMOTE).
- To avoid having overlapping/similar synthetic data, we could use SMOTE with downsampling techniques that will eliminate certain data points to get a cleaner synthetic dataset.

```python
from imblearn.combine import SMOTEENN,SMOTETomek
```

```python
smotetomek = SMOTETomek(random_state=0)
X_smotek, y_smotek = smotetomek.fit_resample(X_train, y_train)
plt.title("SMOTE + Tomek Links")
plot_data(X_smotek,y_smotek)
imb_pipeline("SMOTE-TOMEK",smotetomek)
print("The final amount of data:",Counter(y_smotek))
```

- We can see that combining smote+ tomek links reduced the number of samples to 1350 for both classes vs 1450 in the previous algorithms.

```python
smoteENN = SMOTEENN(random_state=0)
X_smotenn, y_smotenn = smoteENN.fit_resample(X_train, y_train)
plt.title("SMOTE Edited nearest neighbor ")
plot_data(X_smotenn,y_smotenn)
imb_pipeline("SMOTEENN",smoteENN)
print("Resampled data count:",Counter(y_smotenn))
```
